importing the important libraries for calling the different models for machine learning and also splitting the dataset 
for training and testing

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,random_state=45,test_size=.25)
```

finding out the best random state

```python
maxac=0
maxrs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=.20,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    accuracy=accuracy_score(y_test,predlr)
    if accuracy>maxac:
        maxac=accuracy
        maxrs=i
print(maxac)
print(maxrs)
```

```python
model=[LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier()]
for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('acc_score=',accuracy_score(y_test,predm))
    print('conf_mat=',confusion_matrix(y_test,predm))
    print('class_report=',classification_report(y_test,predm))
```

```python
model=[RandomForestClassifier()]
for m in model:
    m.fit(x_train,y_train)
    predm=m.predict(x_test)
    print('acc_score=',accuracy_score(y_test,predm))
    print('conf_mat=',confusion_matrix(y_test,predm))
    print('class_report=',classification_report(y_test,predm))
```

```python
kernel_list=['linear','poly','rbf']
for i in kernel_list:
    svc=SVC(kernel=i)
    svc.fit(x_train,y_train)
    svc.score(x_train,y_train)
    predsv=svc.predict(x_test)
    print('acc=',accuracy_score(y_test,predsv))
    print('conf=',confusion_matrix(y_test,predsv))
    print('class=',classification_report(y_test,predsv))
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print('predicted =',predrf)
print('actual =',y_test)
print('accuracy_score=',accuracy_score(y_test,predrf))
print('confusion_matrix=',confusion_matrix(y_test,predrf))
print('classification_report=',classification_report(y_test,predrf))
```

```python
from sklearn.metrics import roc_curve
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
print('predrf',rf.predict_proba(x_test)[:,1])
fpr, tpr, thresholds = roc_curve(y_test,rf.predict_proba(x_test)[:,1],pos_label=True)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```
